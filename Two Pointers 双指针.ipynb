```python
# 1. Two Sum O(n)复杂度
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        nums = [(number,index) for index,number in enumerate(nums)]
        nums = sorted(nums)
        left,right = 0, len(nums) - 1
        while left <= right:
            if nums[left][0] + nums[right][0] > target:
                right -= 1
            elif nums[left][0] + nums[right][0] < target: 
#                 注意这里一定要是elif，否则满足上面条件修改之后还会走这一行，而我们想要的是每次判别一次
                left += 1
            else:
                return [nums[left][1],nums[right][1]]
```

```python
# 125. Valid Palindrome
# 这道题采用相同的two-pointer思想，唯一值得注意的是这次其实并不需要声明enumerate，我只是习惯了这样写了
class Solution:
    def isPalindrome(self, s: str) -> bool:
        """
        :type s: str
        :rtype: bool
        """
        s = s.lower()
        string = [(number,index) for index, number in enumerate(s)]
        left,right = 0, len(string) - 1
        while left < right:
            if string[left][0] == string[right][0]:
                left +=1
                right -=1
            elif not string[left][0].isalpha() and not string[left][0].isdigit():
                left += 1
            elif not string[right][0].isalpha() and not string[right][0].isdigit():
                right -= 1
            else:
                return False
        return True
            
```

```python
# 680. Valid Palindrome II
# 上面那道题的延申，值得注意的是这里要进行一部判断条件的切换
class Solution:
    def validPalindrome(self, s: str) -> bool:
        left,right = self.twopointer(s,0,len(s) - 1)
        if left >= right:
            return True
        return self.changeTrue(s,left +1 , right) or self.changeTrue(s,left, right -1)

    def twopointer(self,s,left,right):
        while left < right:
            if s[left]!= s[right]:
                return left,right
            else:
                left +=1
                right -= 1
        return left,right
    
    def changeTrue(self,s,left,right):
        left,right = self.twopointer(s,left,right)
        if left >= right:
            return True
        return False
```

```python
# 15. 3Sum 
# Three Sum如果用三重for循环来进行计算的话时间复杂度为O(n^3)显然计算较大数据时行不通，那么我们需要在Two Sum基础上
# 进行修改。这道题要求a + b + c = 0, 那我我们可以把这个变成Two Sum问题 a + b = 0 - c = target，我们可以使用一个for
# 循环 + Two Sum来进行计算，注意这样的话for循环复杂度O(N) x Two Sum复杂度O(N) = O(N^2) 简化了计算
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        results = []
        for i in range(0, len(nums) - 2):
#             i最多到len(nums) -2 因为要留最后两个位置分别给left和right
            if i > 0 and nums[i] == nums[i - 1]:
                continue
            self.find_two_sum(nums, i + 1, len(nums) - 1, -nums[i], results)
#             左指针在i循环的下一个
        return results
            
    def find_two_sum(self,nums,left,right,target,results):
        while left < right:
            if nums[left] + nums[right] == target:
                results.append([-target,nums[left],nums[right]])
                left +=1
                right -= 1
                while left < right and nums[left] == nums[left - 1]:
                    left += 1
                while left < right and nums[right] == nums[right + 1]:
                    right -= 1
            elif nums[left] + nums[right] > target:
                right -= 1
            else:
                left += 1
```

```python
# 18. 4Sum
# 4Sum就是在3Sum基础上进行修改,记住for循环的写法就行
class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        results = []
        nums.sort()
        for i in range(0, len(nums)-3):
            if i > 0 and nums[i] == nums[i - 1]:
                continue
            for j in range(i + 1, len(nums)-2):
                if j > i + 1 and nums[j] == nums[j - 1]:
                    continue
                self.two_sum(nums,target - nums[i] - nums[j], j + 1, len(nums) - 1, results,i,j)
        return results
    def two_sum(self,nums,target,left,right,results,i,j):
        while left < right:
            if nums[left] + nums[right] == target:
                results.append([nums[i],nums[j],nums[left],nums[right]])
                left += 1
                right -= 1
                while left < right and nums[left] == nums[left - 1]:
                    left += 1
                while left < right and nums[right] == nums[right + 1]:
                    right -= 1
            elif nums[left] + nums[right] > target:
                right -= 1
            else:
                left += 1
```

```python
# 611. Valid Triangle Number
# 这道题得再看，这个total的写法
class Solution:
    def triangleNumber(self, nums: List[int]) -> int:
                
        nums = sorted(nums)
        total = 0
        for i in range(len(nums)-2):
            if nums[i] == 0:
                continue
            end = i + 2
            for j in range(i+1, len(nums)-1):
                while end < len(nums) and nums[end] < (nums[i] + nums[j]):
                    end += 1 #注意其移动的是end，end是最大边
                total += end - j - 1
        return total
```

```python
# 1099. Two Sum Less Than K
class Solution:
    def twoSumLessThanK(self, A: List[int], K: int) -> int:
        A.sort()
        count = -1
        left,right = 0, len(A) - 1
        while left < right:
            if A[left] + A[right] >= K:
                right -= 1
            elif A[left] + A[right] < K:
                if count < A[left] + A[right]: #跟count进行比较来判断是否进行更新
                    count = A[left] + A[right]
                    left += 1
                else:
                    left += 1
        return count
```

```python
# 16. 3Sum Closest
class Solution:
    def threeSumClosest(self, nums: List[int], target: int) -> int:
        nums.sort()
        ans = None
        for i in range(len(nums)):
            left = i + 1
            right = len(nums) - 1
            while left < right:
                summary = nums[left] + nums[right] + nums[i]
                if summary == target:
                    return target
                elif ans is None or abs(summary - target) < abs(ans - target):
                    ans = summary
                elif summary <= target:
                    left += 1
                else:
                    right -= 1
        return ans
    
时间复杂度
数组排序的时间复杂度为O(nlogn)
遍历过程，固定值为 n 次，双指针为 n 次，时间复杂度为 O(n^2)
总时间复杂度：
O(nlogn)+O(n^2)=O(n^2)

空间复杂度为O（1），只需要常量空间。
```

```python
# 922. Sort Array By Parity II
# 这道题的要求是奇数位置放奇数，偶数位置放偶数
# 这道题同样也是同向双指针
class Solution:
    def sortArrayByParityII(self, A: List[int]) -> List[int]:
        even,odd = 0, 1
        while even< len(A) and odd < len(A):
            while even< len(A) and odd < len(A) and A[even] % 2 == 0:
                even += 2
            while even< len(A) and odd < len(A) and A[odd] % 2 != 0:
                odd += 2
            if A[odd] % 2 == 0 and A[even] % 2 != 0:
                A[even],A[odd] = A[odd],A[even]
        return A
```

```python
# lintcode partition array
# partition是这种将一个list分成几个部分的问题
# partition是类似于这种左右进行交换的问题，这是一个标准模板，记一下
# 非常需要注意的地方是这里的while循环都是小于等于！！！
# 必须要有等于号，否则会出问题
class Solution:
    """
    @param nums: The integer array you should partition
    @param k: An integer
    @return: The index after partition
    """
    def partitionArray(self, nums, k):
        # write your code here
        left, right = 0, len(nums) - 1
        while left <= right:
            while left <= right and nums[left] < k:
                left += 1
            while left <= right and nums[right] >= k:
                right -= 1
            if left < right:
                temp = nums[left]
                nums[left] = nums[right]
                nums[right] = temp
                
                left += 1
                right -= 1
        return left
```

```python
# 与上面这道题一样，这道题也用到了quick sort的方法
# Lintcode Kth Smallest Numbers in Unsorted Array
import random
class Solution:
    """
    @param k: An integer
    @param nums: An integer array
    @return: kth smallest element
    """
    def kthSmallest(self, k, nums):
        # write your code here
        left = []
        right = []
        equal = []
        pivot = random.choice(nums)
        for num in nums:
            if num < pivot:
                left.append(num)
            elif num == pivot:
                equal.append(num)
            else:
                right.append(num)
        
        if len(left) == k - 1: #这个底下的顺序尤其要注意，得先判断len(left) >= k这种情况再判断底下那种
            return pivot
        if len(left) >= k:
            return self.kthSmallest(k, left)
        if len(left) + len(equal) >= k:
            return pivot
        if len(left) + len(equal) < k:
            return self.kthSmallest(k - len(equal) - len(left), right)
```

```python
# 75. Sort Colors 要再写一遍
# 使用一次扫描的办法。 设立三根指针，left, index, right。定义如下规则：

# left 的左侧都是 0（不含 left）
# right 的右侧都是 2（不含 right）
# index 从左到右扫描每个数，如果碰到 0 就丢给 left，碰到 2 就丢给 right。碰到 1 就跳过不管。
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        left, index, right = 0, 0,len(nums) - 1
        while index <= right:
            if nums[index] == 0:
                nums[left],nums[index] = nums[index], nums[left]
                index += 1
                left += 1
            elif nums[index] == 2: #这几个是并列关系，在走之前先通过while判断一下
                nums[right],nums[index] = nums[index], nums[right]
                right -= 1
            else:
                index += 1
        return nums
```

```python
# Rainbow sort
# 彩虹sort，是之前sort colors的升级版，这时候不仅仅只有三个颜色了而是有多个颜色，然后要将其分类放一块，这里用到了
# 分治法，能将复杂度降到O(nlogk)，这是一个two pointers + 分治的想法
# 分治算法(nlogk)与快排(nlogn)时间复杂度是不同的 因为k < n


class Solution:
    """
    @param colors: A list of integer
    @param k: An integer
    @return: nothing
    """
    def sortColors2(self, colors, k):
        # write your code here
        self.sort(colors, 1, k, 0, len(colors) - 1)
    def sort(self, colors, color_from, color_to, index_from, index_to):
        if color_from == color_to or index_from == index_to:
            return
            
        color = (color_from + color_to) // 2
        
        left, right = index_from, index_to
        while left <= right:
            while left <= right and colors[left] <= color:
                left += 1
            while left <= right and colors[right] > color:
                right -= 1
            if left <= right:
                colors[left],colors[right] = colors[right], colors[left]
        self.sort(colors, color_from, color, index_from, right)
        self.sort(colors, color + 1, color_to, left, index_to)
```

```python
# 280. Wiggle Sort
class Solution:
    def wiggleSort(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
# Method 1 复杂度O(n)单次for循环解决
        for i in range(1, len(nums)):
            if (i % 2 == 1) ^ (nums[i] > nums[i - 1]): #^异或，两个都满足返回0，满足一个返回1
                nums[i], nums[i - 1] = nums[i - 1], nums[i]
# Method 2复杂度O(nlogn)，因为使用了一次sort排序，排完序之后调换顺序解决
        nums.sort()
        if len(nums) < 2:
            return nums
        left = 2
        right = 1
        while left < len(nums) and right < len(nums):
            if left < len(nums) and right < len(nums) and nums[right] < nums[left]:
                nums[left],nums[right] = nums[right],nums[left]
                left += 2
                right += 2
            else:
                left += 2
                right += 2
        return nums
```

```python
# 724. Find Pivot Index这是一个单指针的题
# 这道题是找左边的和等于右边的和
# 值得注意的一个地方是因为L_sum += nums[i + 1]，所以i的取值范围是i < len(nums) - 1
# 当存在正好最后一个即左边的所有和等于右边唯一一个数和的时候时，需要在最后额外判断一句
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        if len(nums) == 0:
            return -1
        i = 0
        L_sum = sum(nums[0:i+1])
        R_sum = sum(nums[i:])
        while i < len(nums) - 1:
            if L_sum == R_sum:
                return i
            else:
                L_sum += nums[i+1]
                R_sum -= nums[i]
                i += 1
        if L_sum == R_sum: #最后额外判断，不然的话上述情况就会返回-1   [-1,-1,0,1,1,0]
            return i
        return -1 
```

```python
# Lintcode Sort Letters by Case
# 将字母大写小写分开成两个部分
# 这明显就是一道partition的题
class Solution:
    """
    @param: chars: The letter array you should sort by Case
    @return: nothing
    """
    def sortLetters(self, chars):
        # write your code here
        if len(chars) == 1:
            return chars
        left,right = 0, len(chars) - 1
        while left <= right:
            while left <= right and chars[left].islower():
                left += 1
            while left <= right and chars[right].isupper():
                right -= 1
            if left <= right:
                chars[left],chars[right] = chars[right],chars[left]
                left += 1
                right -= 1
        return chars
```

```python
# 83. Remove Duplicates from Sorted List
# 这道题唯一要值得注意的时这是个链表形式的，注意写法就行
class Solution:
    def deleteDuplicates(self, head: ListNode) -> ListNode:
        p = head
        while(p != None and p.next != None):
            if p.val != p.next.val:
                p = p.next
            else:
                p.next = p.next.next
        return head
```

```python
# 141. Linked List Cycle
# 这是一道链表循环的题，注意题目中的pos是不可读的，通过一个快慢指针迟早相遇
# 快慢指针的经典题。
# 快指针每次走两步，慢指针一次走一步。 在慢指针进入环之后，快慢指针之间的距离每次缩小1，所以最终能相遇。
# 注意这个相遇并不意味着指向圆环的入口，他有可能实在环内的某一位置相遇
class Solution:
    def hasCycle(self, head: ListNode) -> bool:
        if head is None:
            return False
        headA,headB = head, head
        while True:
            if headA.next is not None:
                headA = headA.next.next #A走的更远所以上面用headA.next来判断下一个存不存在
                headB = headB.next
                if headA is None or headB is None:
                    return False
                elif headA == headB:
                    return True
            else:
                return False
        return False
```

```python
# 142. Linked List Cycle II
# 这道题与前面一题不一样的是他要找到进入循环的位置，
class Solution:
    def detectCycle(self, head: ListNode) -> ListNode:
        if not head or head.next == None:
            return None
        headA = head
        headB = head
        while headA is not None and headA.next is not None:
            headA = headA.next.next
            headB = headB.next
#             这里是找到第一次相遇位置
            if headA == headB:
                break
        if headA == headB:
            headB = head#让慢的那边重新回到头部
            while headA != headB:
                headA = headA.next
                headB = headB.next
            if headA == headB:#他们俩再次相会时候的位置就是进入循环的位置
                return headB
        return None
```

```python
# 287. Find the Duplicate Number
# 这道题是可以用快慢指针的方法来做的，这里使用了O(nlogn)的二分查找法+鸽笼原理来做
# 如果使用类似于链表的快慢指针其复杂度为O(n)
class Solution:
    def findDuplicate(self, nums: List[int]) -> int:
        start,end = 0, len(nums) - 1
        while start + 1 < end:
            mid = (start + end)//2
            if self.smaller_or_equal(nums,mid) > mid:
                end = mid
            else:
                start = mid
        if self.smaller_or_equal(nums,start) > start:
            return start
        return end
    def smaller_or_equal(self,nums,val):
        count = 0
        for num in nums:
            if num <= val:
                count += 1
        return count
# 二分枚举答案范围，使用鸽笼原理进行检验
# 根据鸽笼原理，给定n + 1个范围[1, n]的整数，其中一定存在数字出现至少两次。
# 假设枚举的数字为 n / 2：
# 遍历数组，若数组中不大于n / 2的数字个数超过n / 2，则可以确定[1, n /2]范围内一定有解，
# 否则可以确定解落在(n / 2, n]范围内。
            
```

```python
# 283. Move Zeroes
# 题目要求in-place操作,因此不能额外申明空间来进行
# 这道题我的想法是使用快慢指针来做，也不能完全叫快慢指针吧，就是
# 一个指针遇到零时停下，另一个指针一直向前找到第一个非零的这时候这俩进行交换
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        if len(nums) <= 1:
            return nums
        slow,fast = 0,0
        while fast <= len(nums) - 1:
            while fast <= len(nums) - 1 and nums[slow]!= 0:
                slow += 1
                fast += 1
            if fast <= len(nums) - 1 and nums[slow] == 0:
                fast += 1
                while fast <= len(nums) - 1 and nums[fast] == 0:
                    fast += 1
                if fast <= len(nums) - 1 and nums[fast] != 0:
                    nums[slow],nums[fast] = nums[fast],nums[slow]
                    slow += 1
        return nums
```

```python
前面两种Two Pointers问题的总结：
不需要维持相对顺序vs需要维持相对顺序 算法上的区别
    不需要维护相对顺序，可以使用partition的方法，其时间复杂度和交换次数都是最优的
    需要维护相对顺序的，只能使用同向双指针的方法，时空复杂度最优，但是交换次数不是最优
```
