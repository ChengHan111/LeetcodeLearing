```python
# 采用的是BFS非递归的方式
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
from collections import deque
class Solution:
    def levelOrder(self, root: TreeNode) -> List[List[int]]:
        if root is None:
            return []
        queue = deque([root]) #deque 既可以pop也可以popleft，pop出来是最后一个数，popleft是第一个数
        #这里实际上是deque了所有root中的数据，root包含了整个二叉树的所有关系
        result = [] #result是最后的返回结果，level是暂存结果
        while queue:
            level = []
            for i in range(len(queue)):
                node = queue.popleft() #node得出的是treenode此时的节点和该节点下其他的数 且pop出来的是最先进去的值，先入先出
                print(node) #对于一开始最上面的root，queue.popleft把整个树都pop出来了
                level.append(node.val) #node.val返回的是该节点的值
                if node.left (is not None可以写，is true返回错误):
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            result.append(level) #result一层一层的append level
        return result  #同理对于反向的二叉树，也可以这样导出result[::-1]就行
```

```python
# 1302. Deepest Leaves Sum
# 由上面这题延申出来的题，算最深层次叶子的和，其实这道题bfs，dfs都可以做，我们这里展示了bfs的思想，因为比较合理
# 因为bfs是层级遍历，最后一层直接求和就可以了。
class Solution:
    def deepestLeavesSum(self, root: TreeNode) -> int:
        if not root:
            return []
        queue = collections.deque()
        queue.append(root)
        result = []
        while queue:
            level = []
            for i in range(len(queue)):
                node = queue.popleft()
                level.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            result.append(level)
        return sum(result[-1])
```

```python
# 101. Symmetric Tree 对称二叉树问题，主要考察的是这三个判断情况如何判断
from collections import deque
class Solution:
    def isSymmetric(self, root: TreeNode) -> bool:
        if root is None:
            return True
        dq = collections.deque([(root, root)]) #相当于用了两个binary-tree一起判断
        while dq:
            #注意这里的判断只是用来判断val这个值的，及上面节点，底下的左右节点append之后在queue的下一步轮回判断，那时候val就是这一行下去的left，right了
            node1, node2 = dq.popleft()
            if not node1 and not node2:#如果两个点都是null，那么它们是相等的。返回true
                continue
            if not node1 or not node2:#如果一个点是null，另外一个不是null，那么它们不相等，返回false 
                return False
            if node1.val != node2.val:#如果两个点都不是null，但是它们的值不相等， 返回false 
                return False
            # node1.left and node2.right are symmetric nodes in structure
            # node1.right and node2.left are symmetric nodes in structure
            dq.append((node1.left,node2.right)) #这个append也很精髓
            dq.append((node1.right,node2.left)) #这两个append都要写
        return True
```

```python
# 133. Clone Graph
# Definition for a Node.
class Node: #注意底下调的class Node
    def __init__(self, val = 0, neighbors = []):
        self.val = val
        self.neighbors = neighbors
class Solution:
    def cloneGraph(self, node: 'Node') -> 'Node':
        # return copy.deepcopy(node)
        root = node
        if node is None:
            return node #这里不是return None！是return的node
            
        # use bfs algorithm to traverse the graph and get all nodes.得到所有的nodes
        nodes = self.getNodes(node)
        
        # copy nodes, store the old->new mapping information in a hash map
        mapping = {}
        for node in nodes: #要用一个for循环来map所有的node
            mapping[node] = Node(node.val) #把这个映射关系存起来了
        
        # copy neighbors(edges)
        for node in nodes:
            new_node = mapping[node] #把存起来的映射关系赋给新的node
            for neighbor in node.neighbors:
                new_neighbor = mapping[neighbor]
                new_node.neighbors.append(new_neighbor)#复制到new_node.neighbors里面
        
        return mapping[root]
        
# BFS是一种思想,是一种在自定义函数中规格非常接近的思想
    def getNodes(self, node): #The getNodes is used to prevent cycles and get the cloned copy of a node.
        q = collections.deque([node])
        result = set([node])
        while q:# append neighbor之后，q刚开始没有值了，因为popleft了，append之后可能会有多值，取决于neighbor的数量
            head = q.popleft() #这里赋的一个head值，然后在下面调用其neighbor
            for neighbor in head.neighbors: #前面没有声明过为什么可以直接head.neighbors
                if neighbor not in result:
                    result.add(neighbor) #set 通过add来添加元素
                    q.append(neighbor)
        return result
```

```python
# 126. Word Ladder II
# Word Ladder 的第二题，这个与第一题的不同是第一题只要求找到一个最短路径。
# 通过BFS可以找到最短路径，这里不同的是我要找到所有的最短路径，显然用BFS就不可以了，因为BFS只能找到其中一个解，
# 但是BFS是可以找到最短的distance的，即值，找到值之后再通过DFS对所有的路径进行搜索，这样就可以找到所有的可能
# 有一个非常值得注意的是这里设置的dictionary来记住到达每一层的距离，而这个到达每一层的最短距离是固定的，因此在
# 进行DFS的时候可以进行剪枝
from collections import deque
class Solution:
    def findLadders(self, beginWord, endWord, wordList):
        wordList = set(wordList)
        wordList.add(beginWord)
        # wordList.add(endWord) wordList中包含了endWord了，不用管
        distance = {}
        result = []
        self.bfs(wordList, beginWord, distance)
        
        print(distance)
        
        self.dfs(distance, beginWord, [beginWord], result, endWord, beginWord, wordList)
        return result

    def bfs(self, wordList, beginWord, distance):
        distance[beginWord] = 0
        queue = deque([beginWord])
        while queue:
            word = queue.popleft()
            for next_word in self.get_next_word(word, wordList):
                if next_word not in distance:
                    distance[next_word] = distance[word] + 1
                    queue.append(next_word)

    def get_next_word(self, word, wordList):
        words = []
        for i in range(len(word)):
            for char in 'abcdefghijklmnopqrstuvwxyz':
                if word[i] != char:
                    next_word = (word[:i] + char + word[i + 1:])
                    if next_word != word and next_word in wordList:
                        words.append(next_word)
        return words

    def dfs(self, distance, beginWord, current, result, endWord, curtWord, wordList):
        if curtWord == endWord:
            result.append(list(current))
            return

        for word in self.get_next_word(curtWord, wordList):
            if distance[word] != distance[curtWord] + 1:
                continue
            current.append(word)
            self.dfs(distance, beginWord, current, result, endWord, word, wordList)
            current.pop()
```

```python
# 200. Number of Islands
# string 和 Boolean的不同,string是字符串型的,而boolean是布尔型的,只能返回 True False
from collections import deque
class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        if grid is None: #or not grid[0]:
            return 0
        
        islands = 0
        for i in range(len(grid)):#rows
            for j in range(len(grid[0])):#columns
                if grid[i][j] == '1':
                    self.bfs(grid, i, j) #找周围区域看是否有存在1的位置，共同作为一个island
                    islands += 1
                    
        return islands                    
    
    def bfs(self, grid, x, y):
        #往下这两行十分重要
        queue = deque([(x, y)]) #注意这里deque的是一组(x,y) 用括号包起来的
        grid[x][y] = '0' #初始化,把这个给翻过来了，不然的话到时候周边搜索的时候搜到他又是‘1’，不就死循环了
        while queue:
            x, y = queue.popleft()
            for delta_x, delta_y in [(1, 0), (0, -1), (-1, 0), (0, 1)]: #四方向写法
                next_x = x + delta_x
                next_y = y + delta_y
                if self.is_valid(grid,next_x,next_y):
                    queue.append((next_x,next_y)) #queue append了，相当于我搜索区域变多了现在，
#                   是一个逐步满足逐步扩大的过程,注意双括号，因为append的是一个整体的(next_x,next_y)
                    grid[next_x][next_y] = '0' #做完了is_valid才翻过来
#                 if not self.is_valid(grid, next_x, next_y):#二维时,调用了新的函数
#                     continue
#                 queue.append((next_x, next_y))
#                 grid[next_x][next_y] = '0' #这里归零用来走for循环的下一个，相当于清零一次
                
    def is_valid(self, grid, x, y):
        n, m = len(grid), len(grid[0])
        if 0 <= x < n and 0 <= y < m and grid[x][y] == '1':
            return grid
```

```python
# 286. Walls and Gates
#相当于进去要先append，相当于先进去的会先搜索，因此distance会找到最短distance，后面加进来的距离都append到后头了还没pop出来，deque的好处就在这里
# 体现出来了。而在前一题岛屿问题的时候并不需要知道谁先谁后，只要能够框出区域就行。
from collections import deque
class Solution:
    def wallsAndGates(self, rooms: List[List[int]]) -> None:
        """
        Do not return anything, modify rooms in-place instead.
        """
        if not rooms: #注意一定要这么写，因为输入None是一种情况，输入[]是一种情况
            return []
        # if rooms is None:
        #     return []
        r_size  = len(rooms)
        c_size = len(rooms[0])
        queue = collections.deque()
        for i in range(r_size):
            for j in range(c_size):
                if rooms[i][j] == 0: #这道题和island题不同的一点在于这里要先append所有的为0的点的情况，不然会出现距离不是最短的情况。
                    queue.append((i,j)) 
        while queue:
            x,y = queue.popleft()
            distance = rooms[x][y]+1  #注意这两个distance写的位置
            direction = [(-1,0),(1,0),(0,-1),(0,1)]
            for delta_x,delta_y in direction:
                next_x = x + delta_x
                next_y = y + delta_y
                if 0<=next_x<r_size and 0<=next_y<c_size and rooms[next_x][next_y] == 2147483647:
                    rooms[next_x][next_y] = distance
                    queue.append((next_x,next_y))
        return rooms
```

```python
# 236. Lowest Common Ancestor of a Binarmay Tree 罗列二叉树所有LCA所有情况的题
# Definition for a binary tree node.
# class TreeNode:
#  https://leetcode.com/problems/lowest-common-ancestor-of-a-binary-tree/   def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
from collections import deque
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if root and ( root is p or root is q ): #首先我们要想大的类型，在root，在左右分叉树，没有这三种情况
			# hit
			# root is either node p or node q
            return root
        
        if root is None:
			# empty tree or empty node
            return None
            
        else:#注意，几个大分类要在外面，这些属于左右二叉树的小分类，写在内部
			# common ancestor of p, q exists in left sub-tree
            left_ancestor = self.lowestCommonAncestor( root.left, p ,q) #注意，这里的自身调用返回的是root.left而不是函数中的root，因为你已经在left tree 中进行的选择
             
			# common ancestor of p, q exists in right sub-tree
            right_ancestor = self.lowestCommonAncestor( root.right, p ,q)
            
            if left_ancestor and right_ancestor:
				# p, q reside in two sides, one in left sub-tree, the other in right sub-tree
                return root
            
            elif left_ancestor:
				# both p, q reside in left sub-tree
                return left_ancestor
            
            elif right_ancestor:
				# both p, q reside in right sub-tree
                return right_ancestor
            
            else:
				# both p, q do not exist in current binary tree
                return None
```

```python
# 拓扑排序 Topological Sorting 这种拓扑排序的方法也可以用bfs来解决
# 几乎每个公司都有一道拓扑排序的题
# 注意拓扑排序主要针对的就是有向无环图（树也包含），如果有环的时候无法判断先后顺序
# 入度（In-degree）：有向图（Directed Graph）中指向当前节点的点的个数（或指向当前节点的边的条数）
# 算法描述：
# 1. 统计每个点的入度
# 2. 将每个入度为 0 的点放入队列（Queue）中作为起始节点
# 3. 不断从队列中拿出一个点，去掉这个点的所有连边（指向其他点的边），其他点的相应的入度 - 1
# 4. 一旦发现新的入度为 0 的点，丢回队列中
# 拓扑排序并不是传统的排序算法
# 一个图可能存在多个拓扑序（Topological Order），也可能不存在任何拓扑序

class Solution:
    """
    @param graph: A list of Directed graph node
    @return: Any topological order for the given graph.
    """
    def topSort(self, graph):
        node_to_indegree = self.get_indegree(graph) #入度函数

        # bfs
        order = []
        start_nodes = [n for n in graph if node_to_indegree[n] == 0] #将每个入度为0的点放入队列Queue中作为起始节点
        queue = collections.deque(start_nodes)#上面已经是一个list了，那么我们做deque的时候不需要进行类型转换
        while queue:
            node = queue.popleft()
            order.append(node)
            for neighbor in node.neighbors:
                node_to_indegree[neighbor] -= 1 #拿出node之后其他点的相应入度-1
                if node_to_indegree[neighbor] == 0: #如果入度为0了，那么就把新的入度为0的点放入队列queue中
                    queue.append(neighbor)
                
        return order #返回的是拓扑的连接顺序
    
    def get_indegree(self, graph):
        node_to_indegree = {x: 0 for x in graph} #先初始化所有的入度

        for node in graph:
            for neighbor in node.neighbors:
                node_to_indegree[neighbor] += 1
                
        return node_to_indegree
```

```python
# 有向图的一道题
# 1361. Validate Binary Tree Nodes 左右子树都遍历一遍，用的BFS求解
class Solution:
    def validateBinaryTreeNodes(self, n: int, leftChild: List[int], rightChild: List[int]) -> bool:
        visited = set()
        queue = collections.deque()
        for i in range(n):
            if i not in visited:
                queue.append(i)
                while queue:
                    node = queue.popleft()
                    if leftChild[node] !=- 1:
                        queue.append(leftChild[node])
                        if leftChild[node] in visited:
                            return False
                        else:
                            visited.add(leftChild[node])
                        
                    if rightChild[node] !=- 1:
                        queue.append(rightChild[node])
                        if rightChild[node] in visited:
                            return False
                        else:
                            visited.add(rightChild[node])

                        
        if len(visited) == n-1:
            return True
        return False
```

```python
# 323. Number of Connected Components in an Undirected Graph
class Solution:
    def countComponents(self, n: int, edges: List[List[int]]) -> int:
        ############ bfs version ###############################
#         注意这道题向我们展示了多重函数嵌套的另一种形式，在函数内套函数时，不需要用self进行声明，且函数内的参数
# 可以跨函数传递，值得注意的是，子函数内输入的是其需要变化的量，对于其他不变的量不需要重新进行声明或者复制
        if not edges:
            return n #可能是多个不连通点
        memo_list = collections.defaultdict(list)
        for i, j in edges:
            memo_list[i].append(j)
            memo_list[j].append(i)
        
        def bfs(point_need,visited):
            queue = collections.deque()
            queue.append(point_need)
            while queue:
                node = queue.popleft()
                for nbhd in memo_list[node]:
                    if nbhd not in visited:
                        visited.append(nbhd)
                        queue.append(nbhd)
            return visited
        
        def find_node(visited):
            for i in range(n):
                if i not in visited:
                    return i
            return -1
        
        count = 0
        visited = []
        for i in range(n):
            node = find_node(visited)
            if node != -1:
                count +=1
                visited.append(node)
                visited = bfs(node,visited)
            else:
                return count
```

```python
# 1120. Maximum Average Subtree 二叉树上求值，求路径
class Solution:
    def maximumAverageSubtree(self, root: TreeNode) -> float:
        if not root:
            return 0
        self.res = float('-inf') #先对res进行一个初始化，到时候方便进行替换 也可以用-sys.maxsize 一样
        self.dfs(root)
        return self.res
    
    def dfs(self,root):
        if not root:
            return [0,0] #这俩括号，还有一个是底下那个都不需要！
        left_sum,left_num = self.dfs(root.left)
        right_sum,right_num = self.dfs(root.right)
        
        curr_sum = left_sum + right_sum + root.val
        curr_num = 1 + left_num + right_num
        
        self.res = max(self.res, curr_sum/curr_num)
        return [curr_sum,curr_num]
```

```python
# 226. Invert Binary Tree 这个属于二叉树的结构发生变化
# Method 1 使用BFS进行思考，一层一层的进行判断，最后一定要返回root，否则没有返回值
class Solution:
    def invertTree(self, root: TreeNode) -> TreeNode:
        if not root:
            return None
        queue = collections.deque()
        queue.append(root)
        while queue:
            node = queue.popleft()
            node.left, node.right = node.right, node.left 
            #主要就是上面这行将左右互换
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        return root

# Method 2 使用DFS进行计算，自己调自己
class Solution:
    def invertTree(self, root: TreeNode) -> TreeNode:
        if root:
            root.left,root.right = root.right,root.left
            if root.left:
                self.invertTree(root.left)
            if root.right:
                self.invertTree(root.right)
            return root
        
        else:
            return None
```

```python
# 114. Flatten Binary Tree to Linked List

class Solution:
    def flatten(self, root: TreeNode) -> None:
        self.helper(root)
        """
        Do not return anything, modify root in-place instead.
        """
    def helper(self,root):
        if root is None:
            return None
        
        left_last = self.helper(root.left)
        right_last = self.helper(root.right)
        
        if left_last is not None:
            left_last.right = root.right
            root.right = root.left
            root.left = None
        return right_last or left_last or root 
#     注意这个优先级顺序，必须是right_last,left_last,root!!!
#     the above solutions all have a return at the end of the conquer step
        #移那一步只是把树接起来了， 但是对整个链条来说，链条的尾巴就三种情况: !!
#         右子树不为空就是右子树最后一个，否则就看左子树，不为空就是左子树最后一个， 如果都为空，最后一个就是根节点


#     or ：找真值：
#     如果有一个值为真，or 返回第一个真值；
#     如果所有的值都为假，or 返回最后一个假值。

#     and ：找假值：
#         如果有一个值为假，and 返回第一个假值；
#         如果所有的值都为真，and 返回最后一个真值。

#     特殊情况：
#         当表达式的所有值都是真值，并且运算符不全部一致，
#         那么第一个 or 或 and 演算完后直接返回。

class Solution:
    # @param root, a tree node
    # @return nothing, do it in place
    def flatten(self, root):
        if root == None:
            return
        self.flatten(root.left)
        self.flatten(root.right)
        p = root
        if p.left == None:
            return
        p = p.left
        while p.right:
            p = p.right
        p.right = root.right
        root.right = root.left
        root.left = None
```

```python
# 701. Insert into a Binary Search Tree
# 在BST中插入一个点，重构树的题
# 这属于divide and conquer吗，是recursion算法btw
# 可以这么想他是把这个新加的节点放在TreeNode的最下层的某一个位置，一直到没有值的时候才添加的TreeNode(val)
class Solution:
    def insertIntoBST(self, root: TreeNode, val: int) -> TreeNode:
        if not root:
            return TreeNode(val)
        if val > root.val: #通过比较大小来判断是放在左边还是右边
            root.right = self.insertIntoBST(root.right,val)
        if val < root.val:
            root.left = self.insertIntoBST(root.left,val)
        return root
```

```python
# 270. Closest Binary Search Tree Value
# 寻找BST中离target最近的那个数，这个是自己的写法，最后用一个二分法找最接近的位置
class Solution:
    def closestValue(self, root: TreeNode, target: float) -> int:
        res = []
        # res1 = defaultdict(int)
        def helper(root):
            if root:
                if root.left:
                    helper(root.left)
                res.append(root.val)
                if root.right:
                    helper(root.right)
            return res
        helper(root)
        start = 0
        end = len(res) - 1
        while start + 1 < end:
            mid = start + (end - start) //2
            if res[mid] <= target:
                start = mid
            elif res[mid] > target:
                end = mid
        if abs(res[start] - target) > abs(res[end] - target):
            return res[end]
        return res[start]
# 答案做法，通过上下限来夹，有种二分法的意思
class Solution:
    def closestValue(self, root: TreeNode, target: float) -> int:
        upper = root
        lower = root
        while root:
            if root.val > target:
                upper = root
                root = root.left
            elif root.val < target:
                lower = root
                root = root.right
            else:
                return root.val
        if abs(upper.val - target) > abs(lower.val - target):
            return lower.val
        return upper.val
```

```python
# 144. Binary Tree Preorder Traversal
# 先序遍历，仅仅调整了一下遍历结构就实现了
class Solution:
    def preorderTraversal(self, root: TreeNode) -> List[int]:
        res = []
        def preorder(root):
            if root:
                res.append(root.val)
                if root.left:
                    preorder(root.left)
                if root.right:
                    preorder(root.right)
            return res
        preorder(root)
        return res
```

```python
# 145. Binary Tree Postorder Traversal 标准后序遍历
class Solution:
    def postorderTraversal(self, root: TreeNode) -> List[int]:
        res = []
        def helper(root):
            if root:
                if root.left:#左子树优先遍历
                    helper(root.left)
                if root.right:#接着是右子树
                    helper(root.right)
                res.append(root.val)#最后是root
            return res
        helper(root)
        return res
```

```python
# 590. N-ary Tree Postorder Traversal
# PostOrder的稍微有点不一样的例子，这里使用的不是二叉树了，而是有多分叉结构的树状结构，但是此时的结构其实都是类似的
class Solution:
    def postorder(self, root: 'Node') -> List[int]:
        res = []
        def helper(root):
            if root:
                for c in root.children: #注意先遍历的子树，这里子树因为不分左右子树而是都叫children，因此要写个for
                    helper(c)
                res.append(root.val)#再遍历的树
            return res
        helper(root)
        return res
```

```python
# 108. Convert Sorted Array to Binary Search Tree
# 这道题不同的点在于其定义了一个TreeNode，我们之前没有接触过这种自己定义Node的题，我们可以看到TreeNode实际上是调用
# 了上面的class进行创建的，这道题的解法结合了二分法和递归的思想，每次都进行二分，先二分找到中间位置，然后逐次
# 二分创建一个BST

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def helper(self,nums,start,end):
        if start > end:
            return None
        Mid = (start + end) //2
        Root = TreeNode(nums[Mid])
        Root.left = self.helper(nums,start,Mid-1)
        Root.right = self.helper(nums,Mid+1,end)
        return Root
    def sortedArrayToBST(self, nums: List[int]) -> TreeNode:
        return self.helper(nums,0,len(nums)-1)
```

```python
# 这里介绍一下Morris算法，非常的节省复杂度，推荐背下来
# 这里介绍先序遍历和中序遍历

morris遍历是二叉树遍历算法的超强进阶算法，跟递归、非递归（栈实现）的空间复杂度，
morris遍历可以将非递归遍历中的空间复杂度降为O(1)。从而实现时间复杂度为O(N)，而空间复杂度为O(1)的精妙算法。
morris遍历利用的是树的叶节点左右孩子为空（树的大量空闲指针），实现空间开销的极限缩减。

***morris遍历的实现原则
记作当前节点为cur。

1.如果cur无左孩子，cur向右移动（cur=cur.right）
2.如果cur有左孩子，找到cur左子树上最右的节点，记为mostright
    2.1如果mostright的right指针指向空，让其指向cur，cur向左移动（cur=cur.left）#注意是mostright的right
    2.2如果mostright的right指针指向cur，让其指向空，cur向右移动（cur=cur.right）
实现以上的原则，即实现了morris遍历。

morris遍历的实质
建立一种机制，对于没有左子树的节点只到达一次，对于有左子树的节点会到达两次
```

```python
# Morris的先序遍历，注意和上面的是可以对上的，注意指针指向就是下一步的方向
class Solution:
    def preorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        res = []
        pNode = root
        while(pNode):
            if pNode.left == None:
            #左子树为空的时候直接访问，并将指针移动到右子树的根节点
                res.append(pNode.val)
                pNode = pNode.right
            else:
            #因为左子树不为空，所以我们能找到左子树的最右节点
                tmp = pNode.left
                while(tmp.right and tmp.right != pNode): #这个while是用来指向mostright的
                    tmp = tmp.right
                if tmp.right == None:
                #注意这里当tmp.right ==None时才访问，因为先序遍历中，这是第一次访问根节点，等第二次访问的时候左子树已经访问完了
                    res.append(pNode.val) #对于有左子树的会到达两次，因此这时候append节点
                    tmp.right =pNode
                    pNode = pNode.left
                else:
                # tmp.right ==pNode，表示这是第二次访问根节点了，进行还原操作
                    tmp.right = None
                    pNode = pNode.right
        return res
```

```python
# Morris的中序遍历
class Solution:
    def inorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        res = []
        pNode = root
        while(pNode):
            if pNode.left == None:
                res.append(pNode.val)
                pNode = pNode.right
            else:
                tmp = pNode.left
                while(tmp.right and tmp.right!=pNode):
                    tmp = tmp.right
                if tmp.right == None:
                    tmp.right = pNode
                    pNode = pNode.left
                else:
                    res.append(pNode.val) 
#中序遍历和先序遍历唯一不同位置在这，在第二次访问左子树的时候再append，这时候就是左子树先记录然后再记录根节点了
                    tmp.right =None
                    pNode = pNode.right
        return res
```

```python
# 90. Subsets II 这里和上面不同的地方在于存在重复元素求subsets
class Solution:
    def subsetsWithDup(self, nums: List[int]) -> List[List[int]]:
        results = []
        nums_new = sorted(nums)
        if len(nums_new) == 0:
            return []
        self.dfs(nums_new,0,[],results)
        return results
    def dfs(self,nums_new,index,combination,results):
        results.append(list(combination))
        for i in range(index,len(nums_new)):
#             相当于判断当这一个元素和前一个元素相等时，前一个元素是否有被选到
            if i!=index and nums_new[i-1] == nums_new[i]:
                continue
# 为了去重。举个例子，假设输入是[1, 2, 2, 2, 3]，
# 如果不加判断直接做的话，那么就会出现[1, 第一个2, 第二个2, 3],[1, 第一个2, 第三个2, 3]这样的重复。
# 为了避免这样的重复，可以规定当有多个连续数字出现的时候，只有选了在数组中紧邻的前一个相同数字，才能够选当前数字。
# 比如例子中，只有选了第一个2，才能选第二个2,只有选了二个2，才能选第三个2，
# 这样一来，我们上面提到的[1, 第一个2, 第三个2, 3]这种情况就会被排除掉，从而达到去重效果。
            combination.append(nums_new[i])
            self.dfs(nums_new,i+1,combination,results)
            combination.pop()
```

```python
# 第六章，隐式图深度优先搜索DFS
# 39. Combination Sum
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        results = []
        # 集合为空
        if len(candidates) == 0:
            return results
        # 利用set去重后排序
        candidatesNew = sorted(list(set(candidates)))
        # dfs
        self.dfs(candidatesNew, 0, [], target, results)
        return results

    def dfs(self, candidatesNew, index, current, remainTarget,  results):
        # 到达边界
        if remainTarget == 0:
            return results.append(list(current)) #current若是满足条件，即append变为结果
        # 递归的拆解：挑一个数放入current
        for i in range(index, len(candidatesNew)):
            # 剪枝
            if remainTarget < candidatesNew[i]:
                break 
#当遇到第一个情况[2,2,2]时候,下一个仍然接的是2,这时候break了跳出dfs的套娃，接着走下面那句current.pop()，
#pop出最后一个数。注意这时候我们还在for大循环里，这时候走i = 0走完，大循环for开始走i = 1
            current.append(candidatesNew[i])
            self.dfs(candidatesNew, i, current, remainTarget - candidatesNew[i], results)
#         注意上面dfs那里一定是要输入i的，否则一直是从0开始的话会出现组合情况，如2，2，3和2，3，2这样
            current.pop()
```

```python
# 40. Combination Sum II
# 这道题与上面的变化是这里不可以重复使用了，其实变来变去都是剪枝的问题
class Solution:
    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
        results = []
        if len(candidates) == 0:
            return []
        candidates_new = sorted(candidates)
        self.dfs(candidates_new,0,[],target,results)
        return results
    def dfs(self,candidates_new,index,current,remain,results):
        if remain == 0:
            return results.append(list(current))
        for i in range(index, len(candidates_new)):
#             剪枝
            if candidates_new[i] > remain:
                break
#             剪枝2  与上面subset2的剪枝一样
            if i != index and candidates_new[i-1] == candidates_new[i]:
                continue
            current.append(candidates_new[i])
            self.dfs(candidates_new, i+1, current, remain - candidates_new[i], results)
            current.pop()
```

```python
# 46. Permutations 排列，和顺序是有关的，现在题目要求得到全排列可能性
class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        results = []

        # 如果数组为空直接返回空
        if nums is None:
            return []

        # dfs
        used = [0] * len(nums)
        self.dfs(nums, used, [], results)
        return results

    def dfs(self, nums, used, current, results):

        # 找到一组排列，已到达边界条件
        if len(nums) == len(current):
            # 因为地址传递，在最终回溯后current为空导致results中均为空列表
            # 所以不能写成results.append(current)
            # results.append(current[:])
            results.append(list(current))
            return

        for i in range(len(nums)):
            # i位置这个元素已经被用过
            if used[i]:
                continue
            # 继续递归
            current.append(nums[i])
            used[i] = 1
            self.dfs(nums, used, current, results)
            used[i] = 0
            current.pop()
```

```python
# 51. N-Queens
# 皇后问题，非常经典，对于皇后，每行每列只能存在一个皇后且斜对角(两个斜对角方向上的同一条线上)不能出现皇后
# 这里用到了斜对角的数学知识，(row + column) 和(row - column)如果在不同斜对角线上时，是不相同的，但是若是在
# 相同的线上了就变成相同的了，因此需要用斜对角对是否存在进行判断，相当于判断条件从原来的单个visited变成了
# 三个，我们可以用set()来进行，也可以用我们之前常用的建立一个全零矩阵来进行计算
class Solution:
    def solveNQueens(self, n):    
        result = []
        visited_column = set()
        left_up = set()
        right_up = set()
        self.dfs(n, [], result, visited_column, left_up, right_up)
#         print('!!!',left_up)
#         print('???',right_up)
        return result

    def dfs(self, n, current, result, visited_column, left_up, right_up):
        if n == len(current):
            print(list(current))
            result.append(self.draw(current))
            return

        row = len(current)

        for i in range(n):  # 这里判断的是column
            if self.is_valid(i, visited_column, left_up, right_up, current):
                current.append(i)
                visited_column.add(i)
                left_up.add(row + i)
                right_up.add(row - i)
                self.dfs(n, current, result, visited_column, left_up, right_up)
                visited_column.remove(i)
                left_up.remove(row + i)
                right_up.remove(row - i)
                current.pop()

    def is_valid(self, i, visited_column, left_up, right_up, current):
        row = len(current)
        if i in visited_column:
            return False
        if (row + i) in left_up:
            return False
        if (row - i) in right_up:
            return False
        return True
# draw这段很精髓，不是自己写的要看
    def draw(self, permutation):
        board = []
        n = len(permutation)
        for col in permutation:
            row_string = ''.join(['Q' if c == col else '.' for c in range(n)])
            board.append(row_string)
        return board
```

```python
# 52. N-Queens II
# 这道题只需要返回满足条件的长度就行
class Solution:
    def totalNQueens(self, n: int) -> int:
        result = []
        visited_column = set()
        left_up = set()
        right_up = set()
        self.dfs(n,result,visited_column,left_up,right_up,[])
        # print(result)
        return len(result)
    def dfs(self,n,result,visited_column,left_up,right_up,current):
        if n == len(current):
            result.append(list(current))
            return
        row = len(current)
        for i in range(n): #find column
            if self.is_valid(current,i,visited_column,left_up,right_up):
                current.append(i)
                visited_column.add(i)
                left_up.add(row - i)
                right_up.add(row + i)
                self.dfs(n,result,visited_column,left_up,right_up,current)
                visited_column.remove(i)
                left_up.remove(row - i)
                right_up.remove(row + i)
                current.pop()
    def is_valid(self,current,i,visited_column,left_up,right_up):
        row = len(current)
        if i in visited_column:
            return False
        if (row - i) in left_up:
            return False
        if (row + i) in right_up:
            return False
        return True
```

```python
# 17. Letter Combinations of a Phone Number
# 这道题的注意点就是底下那个for循环还有这个dictionary的构建
# 总的来说是一道传统的dfs题目
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        KEYBOARDS = {
            '2':'abc',
            '3':'def',
            '4':'ghi',
            '5':'jkl',
            '6':'mno',
            '7':'pqrs',
            '8':'tuv',
            '9':'wxyz'
        }
        if not digits:
            return []
        result = []
        self.dfs(KEYBOARDS,result,[],0,digits)
        return result
    def dfs(self,KEYBOARDS,result,current,index,digits):

        if index == len(digits):
            result.append(''.join(list(current)))
            return 
            
        for i in KEYBOARDS[digits[index]]:
            current.append(i)
            self.dfs(KEYBOARDS,result,current,index+1,digits)
            current.pop()
```

```python
# 31. Next Permutation
# 这是一道字典排序的题目

# 其实理解一下就是全排列。举个例子。
# 1,2,3,4 / 1,2,4,3 / 1,3,2,4 / 1,3,4,2 / 1,4,2,3 / 1,4,3,2
# 上面就是所谓的字典排序的顺序。
# 按照题目要求，返回下一个，就是按照上面的顺序到下一个排序。

# 时间复杂度为 O(N) 这种属于for循环叠加而不是for循环套for循环，因此复杂度是多个O(N)的叠加 而不是平方什么的
# 空间复杂度为 O(1) 没有用到额外的空间。

# [1,2,7,4,3,1] ====> [1,3,1,2,4,7]
# 规律就是说 从后向前遍历，找到第一个降序的数字，2。
# 然后继续从后向前遍历，找比这个数大的第一个数，3。
# 之后交换两个数字，变成 [1,3,7,4,2,1]
# 然后把后面的数字 反转，变成 [1,3,1,2,4,7] 即为最终答案。

class Solution:
    def nextPermutation(self, nums):
        """
        Do not return anything, modify nums in-place instead.
        """
        k = 0 #这里令一下最后让k = i是因为python中虽然可以break的时候的i直接进行使用但是不好
        l = len(nums)
        # 从右向左查询第一个小于右邻的元素
        for i in range(l-2,-1,-1):
            if nums[i] < nums[i+1]: #注意是降序排列，但是nums[i+1]代表的是上面例子中的7，i代表的是2
                k = i
                break
        else: # 没有找到，说明为降序排列,这时候反转整个list
            nums = nums.reverse()
            return 
        # 从右向左查询第一个大于nums[k]的元素
        for j in range(l-1,-1,-1):
            if nums[j] > nums[k]:
                break
        # 交换k和j
        nums[j],nums[k] = nums[k], nums[j]
        
#         # i后面的序列进行反转
#         nums[k+1:] = nums[-1:k:-1]
        #注意这里写成sorted也是可以的
        nums[k+1:] = sorted(nums[k+1:])
        print(nums)
        return nums
a = Solution()
print(a.nextPermutation([1,2,7,4,3,1]))

# 有一个需要注意的是如果使用传统的排列方法，然后让其从如[1,2,3]开始进行排列，他也是按照从小到大的顺序进行排列的
# 但是问题是这样子会得到所有的排列会导致超时，DFS尤其要注意超时的问题
```
