```python
04
class Solution:
    def findNumberIn2DArray(self, matrix: List[List[int]], target: int) -> bool:
        if not matrix:
            return False
        row = len(matrix)
        col = len(matrix[0])
        i,j = 0,col - 1
        while i < row and j >= 0:
            if matrix[i][j] > target:
                j -= 1
            elif matrix[i][j] < target:
                i += 1
            else:
                return True
        return False
```

```python
知识点：

前序遍历列表：第一个元素永远是 【根节点 (root)】
中序遍历列表：根节点 (root)【左边】的所有元素都在根节点的【左分支】，【右边】的所有元素都在根节点的【右分支】
算法思路：

通过【前序遍历列表】确定【根节点 (root)】
将【中序遍历列表】的节点分割成【左分支节点】和【右分支节点】
递归寻找【左分支节点】中的【根节点 (left child)】和 【右分支节点】中的【根节点 (right child)】
07 
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
        # DFS approach
        if not preorder:
            return None
        loc = inorder.index(preorder[0]) #在inorder中preorder[0]的位置
        print(loc)
        root = TreeNode(preorder[0])
        root.left = self.buildTree(preorder[1 : loc + 1], inorder[: loc])
        root.right = self.buildTree(preorder[loc + 1:], inorder[loc + 1 :])
        return root
```

```python
10.2 斐波那契同款题目，递归深度过深，因此得使用DP来解决，关系找到的方法是写出了前五个格子所对应的方法数，再找规律
class Solution:
    def numWays(self, n: int) -> int:
        DP = [0] * (n+1)
        if n == 0:
            DP[0] = 1
        else:
            DP[0] = 1
            DP[1] = 1
        for i in range(2,n+1):
            DP[i] = DP[i-1] + DP[i-2]
        return DP[n]%1000000007     
```

```python
14.1 DP problem
class Solution:
    def cuttingRope(self, n: int) -> int:
        memo = [-1] * (n + 1)
        for i in range(2, n + 1):
            for j in range(1, i):
                memo[i] = max(memo[i], j*memo[i-j], j * (i-j))
        return memo[n]
```

```python
# 24 reverse the linked-list 这道题还需要仔细体会，二次仍然不会
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        prev, curr = None, head
        while curr:
            # curr.next,prev,curr = prev,curr,curr.next
            next_node = curr.next
            curr.next = prev
            prev = curr
            curr = next_node
        return prev
```
