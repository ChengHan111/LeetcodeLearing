```python
# Binary search is a famous question in algorithm.

# For a given sorted array (ascending order) and a target number, find the first index of this number in O(log n) time complexity. 这个时间复杂度只能用二分法
# If the target number does not exist in the array, return -1.
# Example
# If the array is [1, 2, 3, 3, 4, 5, 10], for given target 3, return 2.
```

```python
#binary-search的基本模板，背！！！！！！！！！！
class Solution:
    # @param nums: The integer array
    # @param target: Target number to find
    # @return the first position of target in nums, position start from 0 
    def binarySearch(self, nums, target):
        if not nums:
            return -1
            
        start, end = 0, len(nums) - 1
        # 用 start + 1 < end 而不是 start < end 的目的是为了避免死循环 start end缩到2时，start<end 就死循环了
        # 在 first position of target 的情况下不会出现死循环
        # 但是在 last position of target 的情况下会出现死循环
        # 样例：nums=[1，1] target = 1
        # 为了统一模板，我们就都采用 start + 1 < end，就保证不会出现死循环
        while start + 1 < end:
            # python 没有 overflow 的问题，直接 // 2 就可以了 下取整
            # java和C++ 最好写成 mid = start + (end - start) / 2
            # 防止在 start = 2^31 - 1, end = 2^31 - 1 的情况下出现加法 overflow
            mid = (start + end) // 2
            mid = start +(end-start)//2 #推荐这种写法
            
            # > , =, < 的逻辑先分开写，然后在看看 = 的情况是否能合并到其他分支里
            if nums[mid] < target:
                # 写作 start = mid + 1 也是正确的
                # 只是可以偷懒不写，因为不写也没问题，不会影响时间复杂度
                # 不写的好处是，万一你不小心写成了 mid - 1 你就错了
                start = mid
            elif nums[mid] == target:
                end = mid
            else: 
                # 写作 end = mid - 1 也是正确的
                # 只是可以偷懒不写，因为不写也没问题，不会影响时间复杂度
                # 不写的好处是，万一你不小心写成了 mid + 1 你就错了
                end = mid
            #最后一段我写成,替换这个else
            elif nums[mid] > target:
                end = mid
            ###################
                
        # 因为上面的循环退出条件是 start + 1 < end
        # 因此这里循环结束的时候，start 和 end 的关系是相邻关系（1和2，3和4这种） #注意相邻关系
        # 因此需要再单独判断 start 和 end 这两个数谁是我们要的答案
#         如果是找 first position of target 就先看 start，找last position先看 end
# 对于不存在重复元素的list，如此题，则不需要明确start和end的顺序
        if nums[start] == target:
            return start
        if nums[end] == target:
            return end
        
        return -1 #如果上面相邻两个数都不满足，证明不满足，返回-1
```

```python
# Total Occurrence of Target
# first and last position 延申出来的一道题，这道题想让你算出target的在list中的数量，方法是找到头尾
# (而找到头尾的方法就是使用二分法找到left和right position)
class Solution:
    """
    @param A: A an integer array sorted in ascending order
    @param target: An integer
    @return: An integer
    """
    def totalOccurrence(self, nums, target):
        if not nums:
            return 0
        if target < nums[0] or target > nums[-1]:
            return 0
        left = 0
        right = 0
        # write your code here
        start = 0
        end = len(nums) - 1
        while start + 1 < end:
            mid = start + (end - start)//2
            if nums[mid] >= target:
                end = mid
            elif nums[mid] < target:
                start = mid
        if nums[end] == target: 
            left =  end
        if nums[start] == target:
            left =  start
        start = 0
        end = len(nums) - 1
        while start + 1 < end:
            mid = start + (end - start)//2
            if nums[mid] > target:
                end = mid
            elif nums[mid] <= target:
                start = mid
        if nums[start] == target: 
            right =  start
        if nums[end] == target:
            right =  end
        if right == 0:
            if nums[left] == target:
                return 1
            else:
                return 0
        else:
            return right - left + 1
```

```python
# Maximum Number in Mountain Sequence
# 存在一先增后减数列，寻找数列中的最大值(也有可能一开始就在减)，但是只先增(或不增)后减
# 与后一道题进行比较
class Solution:
    """
    @param nums: a mountain sequence which increase firstly and then decrease
    @return: then mountain top
    """
    def mountainSequence(self, nums):
        # write your code here
        if not nums:
            return None
        start = 0
        end = len(nums) - 1
        while start + 1 < end:
            mid = start + (end - start)//2
            if nums[mid] > nums[mid-1]: #与其前一个进行比较，如果增大证明在增数列内
                start = mid
            else:
                end = mid #开始减小的话就在减数列了，这时end = mid
        return max(nums[start],nums[end])
```

```python
# 153. Find Minimum in Rotated Sorted Array
# 一直和最后进行比较，使用和上面一样的图去分析就行，在最后返回的两个值中找最小的就行了
class Solution:
    def findMin(self, nums: List[int]) -> int:
        if not nums:
            return -1
        end, start = len(nums) - 1, 0
        while start + 1 < end:
            mid = start + (end - start) // 2
            if nums[mid] >= nums[end]:
                #注意都是用mid和end来进行比较的，而不是start侧的比较，用start侧比较会有一些情况不满足
                start = mid
            else:
                end = mid
        return min(nums[start],nums[end])
```

```python
# 154. Find Minimum in Rotated Sorted Array II
# 这道题是有重复情况的，在有重复情况时，等号就不能随便取了，等号要单独拿出来用来进行位移
class Solution:
    def findMin(self, nums: List[int]) -> int:
        start, end = 0, len(nums) - 1
        while start + 1 < end:
            mid = start + (end - start) //2
            if nums[mid] > nums[end]:
                start = mid
            elif nums[mid] < nums[end]:
                end = mid
            else:
                end = end - 1 #用来位移
        return min(nums[start],nums[end])
```

```python
# Search in a Big Sorted Array
# Given a big sorted array with non-negative integers sorted by non-decreasing order. The array is so big so that you can not get the length of the whole array directly, and you can only access the kth number by ArrayReader.get(k) (or ArrayReader->get(k) for C++).

# Find the first index of a target number. Your algorithm should be in O(log k), where k is the first index of the target number.

# Return -1, if the number doesn't exist in the array.
# 就是那种无法知道end的位置的题最后使用0(logn)的方法算出来，要求了复杂度
# 既然是logn那么就只能使用二分法来处理，这里节省计算量的地方是他并不是注意扫描的而是跳着的，再接着之后的二分法
# 计算复杂度是O(2*log k )，因此就是O(logk)
class Solution:
    """
    @param reader: An instance of ArrayReader.
    @param target: An integer
    @return: An integer which is the first index of target.
    """
    def searchBigSortedArray(self, reader, target):
        # write your code here
        start = 0
        end = 1
        while reader.get(end) < target:
            start = end
            end = end * 2 #跳着查找的
        while start + 1 < end: # 传统二分
            mid = start + (end - start) // 2
            if reader.get(mid) >= target:
                end = mid
            else:
                start = mid
        if reader.get(start) == target:
            return start
        if reader.get(end) == target:
            return end
        return -1
```

```python
# 658. Find K Closest Elements 
# 这道题的方法是先找到右侧指针，然后再向两面进行查找，判断哪边更符合要求就先添加哪边的元素。
class Solution:
    def findClosestElements(self, arr: List[int], k: int, x: int) -> List[int]:
        right = self.findrightcloest(arr,x)
        left = right -1
        
        results = []
        for _ in range(k):
            if self.leftiscloser(arr,left,right,x):
                results.append(arr[left])
                left -= 1
            else:
                results.append(arr[right])
                right += 1
        return results #如果题目写了preferred 从小到大的话写成下面形式,否则的话乱序
#         return sorted(results)
                
    def findrightcloest(self,arr,x):
        start, end = 0, len(arr) - 1
        while start + 1 < end:
            mid = start + (end - start)//2
            if arr[mid] >= x:
                end = mid
            if arr[mid] < x:
#             else:
                start = mid
#     同上面所述，对于进行左侧和右侧判断的情况时，left左侧 end先， right右侧 start先
#     因此这里我们先判断的start侧
        if arr[start] >= x:
            return start
        if arr[end] >= x:
            return end
        return len(arr) - 1
    
    
    def leftiscloser(self,arr,left,right,x):
        if right > len(arr) - 1: #右侧超过arr的长度了，那么左边肯定是closer
            return True
        if left < 0: #左侧顶到头了，不能再往左的情况
            return False
        return x -  arr[left] <= arr[right] - x #判断哪边更近，注意等于符号是因为更倾向与小的数字
```
